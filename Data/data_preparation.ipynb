```python
import pandas as pd

df1 = pd.read_csv('Data_amazon_products.csv')
df2 = pd.read_csv('new_transactions.csv')
```

```python
df1.head()
```

```python
df1.drop(columns=['author', 'description', 'price','price/currency','seller','seller/address/0','seller/address/1','seller/address/2','seller/address/3','seller/address/4','seller/averageRating','seller/businessName','seller/id','seller/name','seller/phone','seller/reviewsCount','seller/url'], inplace=True)
```

```python
df1.head()
```

```python
df2.head()
```

```python
unique_categories1 = df1['Category'].unique()
unique_categories2 = df2['Category'].unique()

unique_categories = pd.concat([pd.Series(unique_categories1), pd.Series(unique_categories2)]).unique()

print(unique_categories)
```

```python
# A function to categorize based on keywords or patterns
def map_categories(Category):
  if isinstance(Category, str):
    if 'Laptop' in Category or 'Cables' in Category or 'Accessories' in Category or 'Electronic' in Category:
        return 'Electronics'
    elif 'Bulbs' in Category or 'Lighting' in Category or 'Cleaners' in Category or 'Dress' in Category:
        return 'Home'
    elif 'Food' in Category or 'Chocolate' in Category or 'Drinks' in Category or 'Cooking' in Category or 'Snack' in Category or 'Sweets' in Category:
        return 'Grocery'
    else:
        return 'Pharma'
  else:
    return 'Other'

df1['Category'] = df1['Category'].apply(map_categories)
df2['Category'] = df2['Category'].apply(map_categories)
```

```python
df1['Category'] = df1['Category'].replace('Other', 'Pharma')
df2['Category'] = df2['Category'].replace('Other', 'Pharma')
```
